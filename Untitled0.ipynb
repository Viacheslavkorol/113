```python
import matplotlib.pyplot as plt

def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0: return False
    return True

def get_kth_prime(k):
    # Контроль введення [cite: 196]
    if not isinstance(k, int) or k <= 0:
        raise ValueError("Введіть ціле позитивне число k")

    primes = []
    num = 2
    while len(primes) < k:
        if is_prime(num):
            primes.append(num)
        num += 1
    return primes[-1], primes

# Тестові приклади [cite: 195]
for k_val in [5, 10, 15]:
    result, _ = get_kth_prime(k_val)
    print(f"{k_val}-те просте число: {result}")

# Візуалізація [cite: 197]
k_limit = 10
_, all_primes = get_kth_prime(k_limit)
plt.bar(range(1, k_limit + 1), all_primes, color='skyblue')
plt.title(f"Перші {k_limit} простих чисел")
plt.xlabel("Порядковий номер (k)")
plt.ylabel("Значення числа")
plt.show()
```
